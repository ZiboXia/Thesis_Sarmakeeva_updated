```python
import numpy as np
import matplotlib.pyplot as plt

# load data from simulation
par1_0 = np.loadtxt('0/position_particle_1.txt')
par1vel_0 = np.loadtxt('0/velocity_particle_1.txt')
par1_1 = np.loadtxt('1/position_particle_1.txt')
par1vel_1 = np.loadtxt('1/velocity_particle_1.txt')
par1_2 = np.loadtxt('2/position_particle_1.txt')
par1vel_2 = np.loadtxt('2/velocity_particle_1.txt')
```

```python
import pandas as pd
data = pd.read_csv('plot-data-sphere.csv')
velocity = pd.read_csv('plot-data-velocity.csv')
pos = pd.read_csv('plot-data_position_new.csv')
```

```python
# compute offset
#offset = 5 - par1[0, 3]

# plot the data
linienstaerke = 2
MarkerGroesse = 8
plt.figure(1)
#plt.plot(par1[:, 0], offset + par1[:, 3], '*',
plt.plot(par1_0[:, 0][:600],par1_0[:, 3][:600], '--',color ='red',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(par1_1[:, 0][:600],par1_1[:, 3][:600], '--',color ='blue',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(par1_2[:, 0][:600],par1_2[:, 3][:600], '--',color ='green',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(pos['x'],pos[' y'], '-',color ='black',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.xlabel('time (s)')
plt.ylabel('position (cm)')
plt.title('Comparison of the z-position of falling particle', fontsize=15)
plt.legend(['Experiment [3]_0','Experiment [3]_1', 'Experiment [3]_2','Resolved CFD-DEM'])
#plt.axis([0, 0.25, 0, 6])
plt.tick_params(labelsize=14)

plt.show()
```

```python
plt.figure(2)
plt.plot(par1vel_0[:,0], np.sqrt(par1vel_0[:, 1]**2 + par1vel_0[:, 2]**2 + par1vel_0[:, 3]**2), '--',color = 'red',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(velocity['x'],velocity[' y'],'-',color ='black',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.xlabel('time (s)')
plt.ylabel('m/s')
plt.title('Comparison of the velocity of a particle', fontsize=15)
plt.legend(['Experiment [3]_0','Experiment [3]_1', 'Experiment [3]_2','Resolved CFD-DEM'])
#plt.axis([0, 0.25, -20, 0])
plt.tick_params(labelsize=14)

plt.show()
```

```python
plt.figure(2)
plt.plot(par1vel_0[:,0][:600], np.sqrt(par1vel_0[:, 1]**2 + par1vel_0[:, 2]**2 + par1vel_0[:, 3]**2)[:600], '--',color = 'red',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(par1vel_1[:, 0][:600], np.sqrt(par1vel_1[:, 1]**2 + par1vel_1[:, 2]**2 + par1vel_1[:, 3]**2)[:600], '--',color = 'blue',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(par1vel_2[:, 0][:600], np.sqrt(par1vel_2[:, 1]**2 + par1vel_2[:, 2]**2 + par1vel_2[:, 3]**2)[:600], '--',color = 'green',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.plot(velocity['x'],velocity[' y'],'-',color ='black',linewidth=linienstaerke, markersize=MarkerGroesse)
plt.xlabel('time (s)')
plt.ylabel('m/s')
plt.title('Comparison of the velocity of a particle', fontsize=15)
plt.legend(['Experiment [3]_0','Experiment [3]_1', 'Experiment [3]_2','Resolved CFD-DEM'])
#plt.axis([0, 0.25, -20, 0])
plt.tick_params(labelsize=14)

plt.show()
```
